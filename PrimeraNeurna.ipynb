```python
import numpy as np

def step(z):
  return 1 if z >= 0 else 0


def neurona (x1, x2, w1, w2, b):
  z=x1*w1 + x2*w2 + b
  y=step(z)
  return z, y

#valores base

w1, w2, b=1, 1, -3 #intenta parecerse a una compuerta AND
print("bias con valor de: " + str(b)+"\n")

casos = [(0,0),(0,1), (1,0),(1,1)]
for x1, x2 in casos:
  z, y = neurona(x1, x2, w1, w2, b)
  print(f"Entrada=({x1}, {x2}) z={z:.1f} salida=> {y}")

#1. Se realiza cambio del valis del bias se probaron varios valores pero solo con -1.5 y -2 la neurona se comportó como AND
#2. Tomando la tendencia de asignar al bias valores cada vez menores (-2.5, -3, -10) las salidas son cero. Con valores cada vez mayores (-1, -0,5, 0, 0.5, 1) las salidas son 1
```
