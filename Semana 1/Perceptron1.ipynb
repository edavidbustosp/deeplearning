```python
#Practica de programación de un perceptrón
import numpy as np

class Perceptron:
  def __init__(self, num_features, learning_rate=0.01, epochs=1000):
    self.weights=np.zeros(num_features+1) # Suma 1 al bias
    self.learning_rate=learning_rate
    self.epochs=epochs

  #función de Activación
  def activation_function(self, x):
    return 1 if x>=0 else 0

  def predict(self, x):
    z=np.dot(x, self.weights[1:])+self.weights[0]
    return self.activation_function(z)

  def train(self, X,y):
    for _ in range(self.epochs):
      for xi, target in zip(X, y):
        update=self.learning_rate * (target - self.predict(xi))
        self.weights[1:] += update * xi
        self.weights[0] += update


#Crear un conjunto de datos simple para la demostración
X=np.array([[0, 0],[0, 1],[1, 0],[1, 1]])
y=np.array([0, 0, 0, 1])

perceptron = Perceptron(num_features=2)
perceptron.train(X,y)

for xi in X:
  print(f"Input: {xi}, Predicted: {perceptron.predict(xi)}")
```
